```
#Use For Loop to print Prime numbers in Between 1to 1000
```

```
 
 
for i in range(1,1000):
 
  if i>0:
 
    for j in range(2,i):
 
        if(i % j==0):
 
            break
 
    else:
 
        print(i)
```

```
#Write a program for Printing the tables from 1 to 10 using Nested For loop
```

```
for i in range(1,11):
  for item in range(1,11):
    print(i,"×",item,"=",item*1)
  print(" ")
```

```
#Write a program to print X prime number using while Loop starting from 0 and take INPUT of X from the user.
```

```
 
x=int(input("please enter a last number: ")) 
a=0
while (a<=x):
    count = 0
    for i in range(2, (a//2 + 1)):
        if(a % i == 0):
            count = count + 1
            break
 
    if (count == 0 and a != 1):
        print(a)
    a=a+1
```
